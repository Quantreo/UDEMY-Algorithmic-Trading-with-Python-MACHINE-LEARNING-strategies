# Section1: Yahoo finance library

```python
# Import yfinance library
import yfinance as yf
```

```python
# Most simple importation
yf.download("GOOG")
```

```python
# Most simple importation
ticker_list = ["GOOG", "EURUSD=X", "^GSPC"]
yf.download(ticker_list)
```

```python
# Personalize date range
yf.download(ticker_list, start="2000-01-01", end="2010-01-01")
```

```python
# Personalize data interval
yf.download(ticker_list, start="2000-01-01", end="2010-01-01", interval="1mo")
```
